```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
file = "../CompactionSinglePointData.parquet"
df = pd.read_parquet(file)
df.drop(columns=["MainComp Compression radial pressure (MPa)","PreComp Compression radial pressure (MPa)"], inplace=True)
df['density'] = df['Weight']/(np.pi * ((df['Diameter']/2)**2)* df['Thickness'])
df.drop(columns=["Weight","Diameter","Thickness"], inplace=True)
```

```python
df.head()
```

```python
# targets
df_density_Y = df['density']
df_hardness_Y = df['Hardness']
df
```

```python
df_X = df.iloc[:,:-2]
df_X = pd.get_dummies(df_X)
```

```python
df_X
```

```python
from sklearn.model_selection import train_test_split
```

```python
# hold_out_test_set for final testing
X_train, X_test, Y_train, Y_test = train_test_split(df_X, df_hardness_Y, test_size= 0.2, random_state=60)
```

    Use cross-validation on the training set to:

    Tune hyperparameters
    Compare models
    Select the best one

    Also - feature engineering and data tweeks - (normalise, scale, encode) 

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
model_LR = LinearRegression()
```

```python
scaler = StandardScaler()
# X_train_scaled = scaler.fit_transform(X_train)  # -> its changing indices -> problematic for k fold
X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train), index=X_train.index, columns=X_train.columns)
X_test_scaled = pd.DataFrame(scaler.fit_transform(X_test),index=X_test.index, columns=X_test.columns)
```

```python
from sklearn.model_selection import KFold, LeaveOneOut, cross_val_score
from sklearn.metrics import r2_score
```

```python
kf = KFold(n_splits=10, shuffle=True, random_state=10)
```

```python
from sklearn.metrics import make_scorer, mean_squared_error

mse_scorer = make_scorer(mean_squared_error)


score = cross_val_score(model_LR, X_train_scaled,Y_train, cv = kf, scoring=mse_scorer)
print("cv scores :", score)
print("avg score :", np.mean(score))
print("std :", np.std(score))
```

```python
# training set - cross val - X_train, X_test#
scores =[]
for train_index, test_index in kf.split(X_train_scaled):
    print(train_index, test_index)
    x_train, x_test = X_train_scaled.iloc[train_index], X_train_scaled.iloc[test_index]
    y_train, y_test = Y_train.iloc[train_index], Y_train.iloc[test_index]
    model_LR.fit(x_train,y_train)

    y_pred = model_LR.predict(x_test)

    scores.append(r2_score(y_test,y_pred))
    print('------------------------------')
print("MSE scores:", scores)
print("Average MSE:", np.mean(scores))
print("Standard deviation:", np.std(scores))
```

```python
leave_one_out = LeaveOneOut()
scores = cross_val_score(model_LR, X_train_scaled,Y_train, cv = leave_one_out, scoring=mse_scorer)

print("MSE scores:", scores)
print("Average MSE:", np.mean(scores))
print("Standard deviation:", np.std(scores))
```

```python
# lasso Regression - Least Absolute Shrinkage and Selection Operator
```

When predictor variables are highly correlated, LASSO tends to select one predictor from the group and ignore the others, which can lead to unstable and non-unique solutions. This behavior can be problematic in situations where it is important to consider all correlated predictors, as ignoring some may result in loss of valuable information

```python
# lasso
from sklearn.linear_model import Lasso

lasso = Lasso() # alpha is regularisation strength

mse_scores = cross_val_score(lasso,X_train_scaled, Y_train, cv=kf, scoring=mse_scorer)
print('avg mse ->',np.mean(mse_scores))
print('std mse ->',np.std(mse_scores))


lasso = lasso.fit(X_train_scaled,Y_train)

y_pred = lasso.predict(X_test_scaled)


print("mse", mean_squared_error(Y_test, y_pred))
```

```python
# feature selection
coefs = pd.DataFrame(lasso.coef_, columns=['coefs'])
vars = pd.DataFrame(X_test.columns, columns=['Variables'])
# print(vars)

# selected_features = lasso.coef_ != 0
```

```python
df = vars.join(coefs)
df = df.loc[df['coefs'] != 0]
```

```python
df.sort_values(by = 'coefs', key=abs, ascending=False)
```

LASSO may not be the best choice when dealing with highly correlated predictors, potential non-unique solutions, and small data sets.

The Elastic Net combines the properties of both LASSO and Ridge Regression by incorporating both L1 and L2 penalties in the regression objective. 

The Elastic Net is particularly useful when dealing with datasets that have highly correlated predictors. Unlike LASSO, which tends to select one predictor from a group of correlated predictors and discard the others, the Elastic Net can select or shrink together correlated predictors, providing a more balanced and robust model.

Ridge Regression: Ridge Regression uses an L2 penalty, which shrinks coefficients but does not set them to zero. This method is useful for handling multicollinearity but does not perform variable selection, leading to models that include all predictors.

Principal Component Regression (PCR): PCR involves transforming the predictors into a set of uncorrelated components using Principal Component Analysis (PCA) and then performing regression on these components. This method helps in dealing with multicollinearity and reducing dimensionality but can be less interpretable due to the transformation of predictors.

Partial Least Squares Regression (PLSR): PLSR is similar to PCR but aims to find components that not only explain the predictorsâ€™ variance but also have a high correlation with the response variable. This method can be more interpretable than PCR and useful for dealing with multicollinearity. - looks at cov?
